# Evaluation of Random Forest model

This evaluation focuses on the best-performing/most-balanced Random Forest model from notebook 8 (`8e_rf_moderate_tuned.joblib`).

### 1.  Final Evaluation on Test-set and Time Series Cross-Validation


```python
import pandas as pd
import joblib
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score
from sklearn.model_selection import TimeSeriesSplit
from sklearn.metrics import f1_score, recall_score
from sklearn.ensemble import RandomForestClassifier
import numpy as np

# Load final dataset
df = pd.read_csv("../data/engineered_traffic_with_lags_and_weather.csv")
df['timestamp'] = pd.to_datetime(df['timestamp'])
df['hour'] = df['timestamp'].dt.hour
df['weekday'] = df['timestamp'].dt.dayofweek

# Load and merge baseline probabilities
prob_df = pd.read_csv("../results/baseline_probabilities.csv")
df = pd.merge(df, prob_df, on=['road', 'hour', 'weekday'], how='left')
df[['prob_severity_0', 'prob_severity_1', 'prob_severity_2']] = df[
    ['prob_severity_0', 'prob_severity_1', 'prob_severity_2']
].fillna(0)


# Define features and target
feature_cols = [
    'prev_1h_severity', 'prev_2h_severity', 'temperature_2m',
    'precipitation', 'rain', 'snowfall', 'wind_speed_10m',
    'wind_gusts_10m', 'cloud_cover',
    'prob_severity_0', 'prob_severity_1', 'prob_severity_2'
]

target = 'severity_level'

X = df[feature_cols]
y = df[target]

# Load the final saved model
final_model = joblib.load("../models/8e_rf_moderate_tuned.joblib")


#  Evaluate on the full test set (70/30 split reused)
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.3, random_state=42, stratify=y
)

y_pred = final_model.predict(X_test)

print("\nFinal Evaluation on Hold-out Test Set")
print("Accuracy:", round(accuracy_score(y_test, y_pred), 4))
print("\nClassification Report:\n", classification_report(y_test, y_pred))
print("\nConfusion Matrix:\n", confusion_matrix(y_test, y_pred))


print("Time Series Cross-Validation")
print("\n============================")

tscv = TimeSeriesSplit(n_splits=5)

accuracies = []
f1_scores = []
recalls = []

for fold, (train_idx, test_idx) in enumerate(tscv.split(X), 1):
    X_train_cv, X_val_cv = X.iloc[train_idx], X.iloc[test_idx]
    y_train_cv, y_val_cv = y.iloc[train_idx], y.iloc[test_idx]

    model = RandomForestClassifier(
        n_estimators=200,
        max_depth=10,
        min_samples_leaf=2,
        min_samples_split=2,
        class_weight="balanced_subsample",
        random_state=42
    )
    model.fit(X_train_cv, y_train_cv)
    y_pred_cv = model.predict(X_val_cv)

    acc = accuracy_score(y_val_cv, y_pred_cv)
    f1 = f1_score(y_val_cv, y_pred_cv, average='macro')
    rec = recall_score(y_val_cv, y_pred_cv, average=None)

    accuracies.append(acc)
    f1_scores.append(f1)
    recalls.append(rec)

    print(f"Fold {fold}")
    print(f"  Accuracy: {acc:.4f}")
    print(f"  Macro F1-score: {f1:.4f}")
    print(f"  Recall per class: {rec}")
    print()

print("Average Results:")
print(f"Mean Accuracy: {np.mean(accuracies):.4f}")
print(f"Mean Macro F1-score: {np.mean(f1_scores):.4f}")
print(f"Mean Recall per class: {np.mean(recalls, axis=0)}")

```

**Results interpretation**

- Accuracy: 0.638 - This shows that the model correctly predicted severity levels 63.8% of the time on unseen data.

- The model performs very well on Class 0 (majority class), but that’s expected.

- Its recall on Class 2 (43%) — far higher than any other model so far (including XGBoost). This means it captures a larger share of severe delays, which are rare but critical.

- This model sacrifices some overall accuracy to perform better on minority classes, which aligns with a fairness-oriented objective.



### 2. Temporal evaluation (e.g. time-aware cross-validation)

Temporal evaluation using TimeSeriesSplit
This involves checking how the model performs over time, which is more realistic for traffic forecasting. 

```python
import pandas as pd

# Load dataset
df = pd.read_csv("../data/engineered_traffic_with_lags_and_weather.csv")
df['timestamp'] = pd.to_datetime(df['timestamp'])
df['hour'] = df['timestamp'].dt.hour
df['weekday'] = df['timestamp'].dt.dayofweek
df['is_rush_hour'] = df['hour'].isin([7, 8, 9, 16, 17, 18]).astype(int)

# Load baseline probabilities
prob_df = pd.read_csv("../results/baseline_probabilities.csv")
prob_df.rename(columns={
    'p_good': 'prob_severity_0',
    'p_minor': 'prob_severity_1',
    'p_serious': 'prob_severity_2'
}, inplace=True)

# Merge
df_merged = pd.merge(df, prob_df, on=['road', 'hour', 'weekday'], how='left')
df_merged[['prob_severity_0', 'prob_severity_1', 'prob_severity_2']] = df_merged[
    ['prob_severity_0', 'prob_severity_1', 'prob_severity_2']
].fillna(0)



from sklearn.model_selection import TimeSeriesSplit
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, f1_score, recall_score, classification_report
import numpy as np

# Prepare data again (to ensure X and y are  defined here)
feature_cols = [
    'prev_1h_severity', 'prev_2h_severity', 'temperature_2m',
    'precipitation', 'rain', 'snowfall', 'wind_speed_10m',
    'wind_gusts_10m', 'cloud_cover',
    'prob_severity_0', 'prob_severity_1', 'prob_severity_2'
]

target = 'severity_level'

X = df_merged[feature_cols].reset_index(drop=True)
y = df_merged[target].reset_index(drop=True)

# Initialise TimeSeriesSplit
tscv = TimeSeriesSplit(n_splits=5)
fold = 1

accuracy_scores = []
f1_scores = []
recall_scores = []

for train_idx, test_idx in tscv.split(X):
    X_train, X_val = X.iloc[train_idx], X.iloc[test_idx]
    y_train, y_val = y.iloc[train_idx], y.iloc[test_idx]

    model = RandomForestClassifier(
        n_estimators=200,
        max_depth=10,
        min_samples_leaf=2,
        min_samples_split=2,
        class_weight='balanced_subsample',
        random_state=42
    )
    model.fit(X_train, y_train)
    y_pred = model.predict(X_val)

    acc = accuracy_score(y_val, y_pred)
    f1 = f1_score(y_val, y_pred, average='macro')
    recall = recall_score(y_val, y_pred, average='macro')

    print(f"\nFold {fold}")
    print("Accuracy:", round(acc, 4))
    print("F1 (macro):", round(f1, 4))
    print("Recall (macro):", round(recall, 4))
    print("Classification Report:\n", classification_report(y_val, y_pred, zero_division=0))

    accuracy_scores.append(acc)
    f1_scores.append(f1)
    recall_scores.append(recall)

    fold += 1

# Summary
print("\nTemporal Cross-Validation Summary")
print("Avg Accuracy:", round(np.mean(accuracy_scores), 4))
print("Avg F1 (macro):", round(np.mean(f1_scores), 4))
print("Avg Recall (macro):", round(np.mean(recall_scores), 4))
```

**Results interpretation**


I used TimeSeriesSplit to mimic a real-world temporal forecasting setup.


| Metric         | Value                                                                                        | Notes                                                          |
| -------------- | -------------------------------------------------------------------------------------------- | -------------------------------------------------------------- |
| Accuracy       | 0.58                                                                                       | Lower than random split, as expected in time-based validation. |
| F1 (macro)     | 0.43                                                                                       | Reflects moderate performance across all classes.              |
| Recall (macro) | 0.55                                                                                        | Good at *capturing positives*, even in minority classes.       |
| Class 0 (Good) | Consistently high precision but lower recall in time-based splits.                           |                                                                |
| Class 1 & 2    | Recall improves in time series vs. random — suggesting model captures rare events over time. |                                                                |



This evaluation uses sliding-window temporal cross-validation to simulate how the model would generalize to unseen future data. Performance remains stable across folds, with an average accuracy of 57.8%, macro F1 of 0.433, and macro recall of 0.556. This is a notable result given the imbalanced classes, especially class 2.

The model consistently performs well on the majority class (severity 0), but also shows consistent moderate recall for minority classes, particularly class 2 in later folds (reaching up to 70%). This suggests the model generalizes fairly across time and maintains useful sensitivity for rare events, which is crucial in real-world deployment.

## 3. Precision-Recall Curve (Per Class)
Purpose: Understand performance on minority classes (1 & 2) without being misled by class 0 dominating accuracy.

Class 2 is rare... PR curves are more informative than ROC when there's class imbalance.

```python
from sklearn.metrics import precision_recall_curve, average_precision_score
import matplotlib.pyplot as plt
from sklearn.preprocessing import label_binarize

# Binarise targets
y_test_bin = label_binarize(y_test, classes=[0, 1, 2])
y_scores = model.predict_proba(X_test)

for i in range(3):
    precision, recall, _ = precision_recall_curve(y_test_bin[:, i], y_scores[:, i])
    ap = average_precision_score(y_test_bin[:, i], y_scores[:, i])
    plt.plot(recall, precision, label=f"Class {i} (AP={ap:.2f})")

plt.xlabel("Recall")
plt.ylabel("Precision")
plt.title("Precision-Recall Curve per Class")
plt.legend()
plt.savefig("../results/prcurve.png", dpi=300)

plt.show()
```

**Results interpretation:**

This plot illustrates the precision-recall trade-off for each severity class (0 = low, 1 = medium, 2 = high). 

Class 0 (low severity): Performs strongly with a consistently high area under the curve (AUC), indicating confident predictions and good separability.

Class 1 (medium severity): Shows a sharp drop-off, reflecting the difficulty the model has in confidently distinguishing this class from others.

Class 2 (high severity): PR curve is low and flat, confirming that the model struggles the most here — due to both class imbalance and overlap with medium severity features.

This analysis highlights the importance of optimizing for macro F1 or class-specific recall, not just overall accuracy.

### 6. Learning curve 

A learning curve was generated to assess how the final Random Forest model generalizes with increasing data. The plot shows both training and validation F1 scores. This helps diagnose underfitting or overfitting and determine whether additional training data might improve performance.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import learning_curve
from sklearn.metrics import make_scorer, f1_score

# Load and prepare data
df = pd.read_csv("../data/engineered_traffic_with_lags_and_weather.csv")
df['timestamp'] = pd.to_datetime(df['timestamp'])
df['hour'] = df['timestamp'].dt.hour
df['weekday'] = df['timestamp'].dt.dayofweek
df['is_rush_hour'] = df['hour'].isin([7, 8, 9, 16, 17, 18]).astype(int)

# Load baseline probabilities
prob_df = pd.read_csv("../results/baseline_probabilities.csv")
df = pd.merge(df, prob_df, on=['road', 'hour', 'weekday'], how='left')
df[['prob_severity_0', 'prob_severity_1', 'prob_severity_2']] = df[
    ['prob_severity_0', 'prob_severity_1', 'prob_severity_2']
].fillna(0)



# Final feature set (same as final model)
feature_cols = [
    'prev_1h_severity', 'prev_2h_severity', 'temperature_2m',
    'precipitation', 'rain', 'snowfall', 'wind_speed_10m',
    'wind_gusts_10m', 'cloud_cover',
    'prob_severity_0', 'prob_severity_1', 'prob_severity_2'
]



X = df[feature_cols]
y = df['severity_level']

# Learning Curve
rf = RandomForestClassifier(
    n_estimators=200,
    max_depth=10,
    min_samples_leaf=2,
    min_samples_split=2,
    class_weight="balanced_subsample",
    random_state=42
)

train_sizes, train_scores, val_scores = learning_curve(
    rf, X, y,
    cv=3,
    scoring=make_scorer(f1_score, average='macro'),
    train_sizes=np.linspace(0.1, 1.0, 5),
    n_jobs=-1,
    shuffle=True,
    random_state=42
)

# Compute mean/std
train_mean = np.mean(train_scores, axis=1)
train_std = np.std(train_scores, axis=1)
val_mean = np.mean(val_scores, axis=1)
val_std = np.std(val_scores, axis=1)

# -----------------------------
# Plot
plt.figure(figsize=(8, 5))
plt.plot(train_sizes, train_mean, 'o-', label="Training F1 (macro)")
plt.fill_between(train_sizes, train_mean - train_std, train_mean + train_std, alpha=0.2)

plt.plot(train_sizes, val_mean, 'o-', label="Validation F1 (macro)")
plt.fill_between(train_sizes, val_mean - val_std, val_mean + val_std, alpha=0.2)

plt.title("Learning Curve – Random Forest (Final Model)")
plt.xlabel("Training Set Size")
plt.ylabel("F1 Score (macro)")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.savefig("../results/learning_curve_rf_final.png", dpi=300)
plt.show()
```

**Results interpretation:**

Training Score: 
- Starts very high (as expected for Random Forests).
- Slightly decreases as the training set size increases — this is normal as the model sees more variability in data.
- Remains consistently higher than the validation score, indicating the model fits the training data well.

Validation Score:
- Starts low with small training sizes.
- Improves gradually as the training size increases.
- Appears to plateau before reaching the maximum size.

Good signs:
- No major overfitting: There is a gap, but it's not extreme — fairly typical for imbalanced classification.
- Validation performance increases with data: Suggests that using more data helps generalization.

Limitations:
- The gap doesn't fully close — this means the model has some bias (underfitting potential), or it’s limited in capturing the complexity of the task.
- Plateauing could also mean adding more data might have diminishing returns unless model complexity or feature richness increases.



### 7. Class-Specific Feature Distributions (e.g., Precipitation, Cloud Cover, Prev Severity)

- prob_severity_1 & prob_severity_2:
These baseline probability features show clear separation between classes, especially for prob_severity_1, further validating the contribution of the baseline model.

- precipitation, rain, snowfall:
Weather-related features show small but meaningful differences — particularly snowfall and precipitation, which have heavier tails in higher severity levels.

- is_rush_hour:
Rush hour periods are associated with increased severity, as expected. The distribution is skewed for severity levels 1 and 2, reflecting real-world traffic trends.

### 8. SHAP Summary Plot (overall feature importance)

SHAP is useful for showing the most important features overall, across all classes.

It uses the mean absolute SHAP value across all samples and all classes (or whichever is dominant).



```python
import shap
import matplotlib.pyplot as plt

feature_cols = [
    'prev_1h_severity', 'prev_2h_severity', 'temperature_2m',
    'precipitation', 'rain', 'snowfall', 'wind_speed_10m',
    'wind_gusts_10m', 'cloud_cover',
    'prob_severity_0', 'prob_severity_1', 'prob_severity_2'
]
X_shap = X_test[feature_cols]

# Create TreeExplainer and get SHAP values 
explainer = shap.TreeExplainer(model)
shap_values = explainer(X_shap, check_additivity=False)

# SHAP summary plot per clas
for class_idx in range(3):
    print(f"SHAP Summary Plot for Class {class_idx}")
    shap.summary_plot(
        shap_values[..., class_idx],   # explicitly extract class-specific values
        X_shap,
        feature_names=feature_cols,
        show=False
    )
    plt.tight_layout()
    plt.savefig(f"../results/shap_summary_class_{class_idx}.png", dpi=300)
    plt.clf()
```
