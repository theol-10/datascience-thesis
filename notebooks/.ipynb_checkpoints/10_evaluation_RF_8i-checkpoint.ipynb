```python
# üìì Notebook 9: Final Model Evaluation (Saved Model)

# ‚úÖ 1. Load final model and data
import pandas as pd
import joblib
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score
from sklearn.model_selection import TimeSeriesSplit
from sklearn.metrics import f1_score, recall_score
from sklearn.ensemble import RandomForestClassifier
import numpy as np

# Load final dataset
df = pd.read_csv("../data/engineered_traffic_with_lags_and_weather.csv")
df['timestamp'] = pd.to_datetime(df['timestamp'])
df['hour'] = df['timestamp'].dt.hour
df['weekday'] = df['timestamp'].dt.dayofweek

# Load and merge baseline probabilities
prob_df = pd.read_csv("../results/baseline_probabilities.csv")
df = pd.merge(df, prob_df, on=['road', 'hour', 'weekday'], how='left')
df[['prob_severity_0', 'prob_severity_1', 'prob_severity_2']] = df[
    ['prob_severity_0', 'prob_severity_1', 'prob_severity_2']
].fillna(0)

# Add interaction features (used in final model)
df['precip_x_cloud'] = df['precipitation'] * df['cloud_cover']
df['temp_x_wind'] = df['temperature_2m'] * df['wind_speed_10m']
df['rushhour_x_prev1h'] = df['is_rush_hour'] * df['prev_1h_severity']

# Define features and target
feature_cols = [
    'prev_1h_severity', 'prev_2h_severity',
    'temperature_2m', 'precipitation', 'rain', 'snowfall',
    'wind_speed_10m', 'wind_gusts_10m', 'cloud_cover',
    'prob_severity_0', 'prob_severity_1', 'prob_severity_2',
    'precip_x_cloud', 'temp_x_wind', 'rushhour_x_prev1h'
]
target = 'severity_level'

X = df[feature_cols]
y = df[target]

# Load the final saved model
final_model = joblib.load("../models/8i_rf_feature_interactions.joblib")


# ‚úÖ 2. Evaluate on the full test set (70/30 split reused)
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.3, random_state=42, stratify=y
)

y_pred = final_model.predict(X_test)

print("\nüìä Final Evaluation on Hold-out Test Set")
print("Accuracy:", round(accuracy_score(y_test, y_pred), 4))
print("\nClassification Report:\n", classification_report(y_test, y_pred))
print("\nConfusion Matrix:\n", confusion_matrix(y_test, y_pred))


# ‚úÖ 3. Time Series Cross-Validation
print("\n============================")
print("‚è±Ô∏è Time Series Cross-Validation")
print("\n============================")

tscv = TimeSeriesSplit(n_splits=5)

accuracies = []
f1_scores = []
recalls = []

for fold, (train_idx, test_idx) in enumerate(tscv.split(X), 1):
    X_train_cv, X_val_cv = X.iloc[train_idx], X.iloc[test_idx]
    y_train_cv, y_val_cv = y.iloc[train_idx], y.iloc[test_idx]

    model = RandomForestClassifier(
        n_estimators=200,
        max_depth=10,
        min_samples_leaf=2,
        min_samples_split=2,
        class_weight="balanced_subsample",
        random_state=42
    )
    model.fit(X_train_cv, y_train_cv)
    y_pred_cv = model.predict(X_val_cv)

    acc = accuracy_score(y_val_cv, y_pred_cv)
    f1 = f1_score(y_val_cv, y_pred_cv, average='macro')
    rec = recall_score(y_val_cv, y_pred_cv, average=None)

    accuracies.append(acc)
    f1_scores.append(f1)
    recalls.append(rec)

    print(f"Fold {fold}")
    print(f"  Accuracy: {acc:.4f}")
    print(f"  Macro F1-score: {f1:.4f}")
    print(f"  Recall per class: {rec}")
    print()

print("‚úÖ Average Results:")
print(f"Mean Accuracy: {np.mean(accuracies):.4f}")
print(f"Mean Macro F1-score: {np.mean(f1_scores):.4f}")
print(f"Mean Recall per class: {np.mean(recalls, axis=0)}")

```

### Temporal evaluation (e.g. time-aware cross-validation)

Temporal evaluation using TimeSeriesSplit
This involves checking how the model performs over time, which is more realistic for traffic forecasting. We already planned to do it like this:

```python
import pandas as pd

# Load dataset
df = pd.read_csv("../data/engineered_traffic_with_lags_and_weather.csv")
df['timestamp'] = pd.to_datetime(df['timestamp'])
df['hour'] = df['timestamp'].dt.hour
df['weekday'] = df['timestamp'].dt.dayofweek
df['is_rush_hour'] = df['hour'].isin([7, 8, 9, 16, 17, 18]).astype(int)

# Load baseline probabilities
prob_df = pd.read_csv("../results/baseline_probabilities.csv")
prob_df.rename(columns={
    'p_good': 'prob_severity_0',
    'p_minor': 'prob_severity_1',
    'p_serious': 'prob_severity_2'
}, inplace=True)

# Merge
df_merged = pd.merge(df, prob_df, on=['road', 'hour', 'weekday'], how='left')
df_merged[['prob_severity_0', 'prob_severity_1', 'prob_severity_2']] = df_merged[
    ['prob_severity_0', 'prob_severity_1', 'prob_severity_2']
].fillna(0)

# Feature interactions
df_merged['precip_x_cloud'] = df_merged['precipitation'] * df_merged['cloud_cover']
df_merged['temp_x_wind'] = df_merged['temperature_2m'] * df_merged['wind_speed_10m']
df_merged['rushhour_x_prev1h'] = df_merged['is_rush_hour'] * df_merged['prev_1h_severity']


from sklearn.model_selection import TimeSeriesSplit
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, f1_score, recall_score, classification_report
import numpy as np

# Prepare data again (to ensure X and y are defined here)
feature_cols = [
    'prev_1h_severity', 'prev_2h_severity',
    'temperature_2m', 'precipitation', 'rain', 'snowfall',
    'wind_speed_10m', 'wind_gusts_10m', 'cloud_cover',
    'prob_severity_0', 'prob_severity_1', 'prob_severity_2',
    'precip_x_cloud', 'temp_x_wind', 'rushhour_x_prev1h'
]
target = 'severity_level'

X = df_merged[feature_cols].reset_index(drop=True)
y = df_merged[target].reset_index(drop=True)

# Initialize TimeSeriesSplit
tscv = TimeSeriesSplit(n_splits=5)
fold = 1

accuracy_scores = []
f1_scores = []
recall_scores = []

for train_idx, test_idx in tscv.split(X):
    X_train, X_val = X.iloc[train_idx], X.iloc[test_idx]
    y_train, y_val = y.iloc[train_idx], y.iloc[test_idx]

    model = RandomForestClassifier(
        n_estimators=200,
        max_depth=10,
        min_samples_leaf=2,
        min_samples_split=2,
        class_weight='balanced_subsample',
        random_state=42
    )
    model.fit(X_train, y_train)
    y_pred = model.predict(X_val)

    acc = accuracy_score(y_val, y_pred)
    f1 = f1_score(y_val, y_pred, average='macro')
    recall = recall_score(y_val, y_pred, average='macro')

    print(f"\nFold {fold}")
    print("Accuracy:", round(acc, 4))
    print("F1 (macro):", round(f1, 4))
    print("Recall (macro):", round(recall, 4))
    print("Classification Report:\n", classification_report(y_val, y_pred, zero_division=0))

    accuracy_scores.append(acc)
    f1_scores.append(f1)
    recall_scores.append(recall)

    fold += 1

# Summary
print("\nüìä Temporal Cross-Validation Summary")
print("Avg Accuracy:", round(np.mean(accuracy_scores), 4))
print("Avg F1 (macro):", round(np.mean(f1_scores), 4))
print("Avg Recall (macro):", round(np.mean(recall_scores), 4))
```

### Learning curve 

A learning curve was generated to assess how the final Random Forest model generalizes with increasing data. The plot shows both training and validation F1 scores. This helps diagnose underfitting or overfitting and determine whether additional training data might improve performance.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import learning_curve
from sklearn.metrics import make_scorer, f1_score

# -----------------------------
# Load and prepare data
df = pd.read_csv("../data/engineered_traffic_with_lags_and_weather.csv")
df['timestamp'] = pd.to_datetime(df['timestamp'])
df['hour'] = df['timestamp'].dt.hour
df['weekday'] = df['timestamp'].dt.dayofweek
df['is_rush_hour'] = df['hour'].isin([7, 8, 9, 16, 17, 18]).astype(int)

# Load baseline probabilities
prob_df = pd.read_csv("../results/baseline_probabilities.csv")
df = pd.merge(df, prob_df, on=['road', 'hour', 'weekday'], how='left')
df[['prob_severity_0', 'prob_severity_1', 'prob_severity_2']] = df[
    ['prob_severity_0', 'prob_severity_1', 'prob_severity_2']
].fillna(0)

# Create interaction feature
df['rushhour_x_prev1h'] = df['is_rush_hour'] * df['prev_1h_severity']

# Final feature set (same as final model)
feature_cols = [
    'prev_1h_severity', 'prev_2h_severity',
    'temperature_2m', 'precipitation', 'rain', 'snowfall',
    'wind_speed_10m', 'wind_gusts_10m', 'cloud_cover',
    'prob_severity_0', 'prob_severity_1', 'prob_severity_2',
    'precip_x_cloud', 'temp_x_wind', 'rushhour_x_prev1h'
]

# Add missing engineered features (if needed)
df['precip_x_cloud'] = df['precipitation'] * df['cloud_cover']
df['temp_x_wind'] = df['temperature_2m'] * df['wind_speed_10m']

X = df[feature_cols]
y = df['severity_level']

# -----------------------------
# Learning Curve
rf = RandomForestClassifier(
    n_estimators=200,
    max_depth=10,
    min_samples_leaf=2,
    min_samples_split=2,
    class_weight="balanced_subsample",
    random_state=42
)

train_sizes, train_scores, val_scores = learning_curve(
    rf, X, y,
    cv=3,
    scoring=make_scorer(f1_score, average='macro'),
    train_sizes=np.linspace(0.1, 1.0, 5),
    n_jobs=-1,
    shuffle=True,
    random_state=42
)

# Compute mean/std
train_mean = np.mean(train_scores, axis=1)
train_std = np.std(train_scores, axis=1)
val_mean = np.mean(val_scores, axis=1)
val_std = np.std(val_scores, axis=1)

# -----------------------------
# Plot
plt.figure(figsize=(8, 5))
plt.plot(train_sizes, train_mean, 'o-', label="Training F1 (macro)")
plt.fill_between(train_sizes, train_mean - train_std, train_mean + train_std, alpha=0.2)

plt.plot(train_sizes, val_mean, 'o-', label="Validation F1 (macro)")
plt.fill_between(train_sizes, val_mean - val_std, val_mean + val_std, alpha=0.2)

plt.title("Learning Curve ‚Äì Random Forest (Final Model)")
plt.xlabel("Training Set Size")
plt.ylabel("F1 Score (macro)")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.savefig("../results/learning_curve_rf_final.png", dpi=300)
plt.show()
```

Interpretation:

Training Score:
Starts very high (as expected for Random Forests).

Slightly decreases as the training set size increases ‚Äî this is normal as the model sees more variability in data.

Remains consistently higher than the validation score, indicating the model fits the training data well.

Validation Score:
Starts low with small training sizes.

Improves gradually as the training size increases.

Appears to plateau before reaching the maximum size.

‚ûï Good signs:
No major overfitting: There is a gap, but it's not extreme ‚Äî fairly typical for imbalanced classification.

Validation performance increases with data: Suggests that using more data helps generalization.

‚ûñ Limitations:
The gap doesn't fully close ‚Äî this means the model has some bias (underfitting potential), or it‚Äôs limited in capturing the complexity of the task.

Plateauing could also mean adding more data might have diminishing returns unless model complexity or feature richness increases.
